```python
import numpy as np
import pandas as pd
import rasterio
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import ConvLSTM2D, BatchNormalization, Conv3D
from tensorflow.keras.layers import ConvLSTM2D, BatchNormalization, Conv3D, Reshape

import matplotlib.pyplot as plt
```

```python
# Function to plot a raster image with a title
def show_band(data, title='', size=(10, 10), cmap='viridis'):
    plt.figure(figsize=size)
    plt.imshow(data, cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.axis('off')
    plt.show()

# Step 1: Load and Preprocess Static Data with Masking

def load_static_raster(file_path, title):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        masked_data = np.ma.masked_equal(data, src.nodata)  # Mask nodata values
    print(f"Loaded {title} with shape {masked_data.shape}")
    show_band(masked_data, title=title)  # Show the raster with color map
    return masked_data

# Paths to your static datasets
elevation_path = 'snapped_data/snapped_DEM.tif'
# slope_path = 'snapped_data/snapped_slope.tif'
landuse_path = 'snapped_data/snapped_landuse.tif'
```

```python
elevation_data = load_static_raster(elevation_path, "Elevation Data")
# slope_data = load_static_raster(slope_path, "Slope Data")
landuse_data = load_static_raster(landuse_path, "Landuse Data")
```

```python
# Stack the static data into a single array (channels_last)
static_data = np.ma.stack([elevation_data, landuse_data], axis=-1)  # Shape: (height, width, channels)
```

```python
# Normalize static data
scaler = StandardScaler()
static_data_reshaped = static_data.reshape(-1, static_data.shape[-1])
static_data_reshaped_scaled = scaler.fit_transform(static_data_reshaped)
static_data = static_data_reshaped_scaled.reshape(static_data.shape)
```

```python
static_data.shape
for i in range(static_data.shape[-1]):
    min_val = np.min(static_data[..., i])
    max_val = np.max(static_data[..., i])
    print(f"Channel {i} - Min: {min_val}, Max: {max_val}")
```

```python
# Visualization of normalized static data
show_band(static_data[..., 0], "Elevation Data (Normalized)", cmap='terrain')
show_band(static_data[..., 1], "Landuse Data (Normalized)", cmap='RdYlBu')
```

```python
print(f"Static data normalized with shape {static_data.shape}")
```

```python
def load_dynamic_csv(file_path):
    df = pd.read_csv(file_path)
    print(f"Loaded dynamic data from {file_path} with shape {df.shape}")
    return df.values  # Convert to NumPy array
```

```python
precipitation_path = 'calraster/precipitation_data.csv'
precipitation_data =  load_dynamic_csv(precipitation_path)
```

```python
precipitation_data.shape
```

```python
# Reshape dynamic data to match the spatial dimensions of the static data
time_steps = len(precipitation_data)
height, width = elevation_data.shape
```

```python
dynamic_data = np.zeros((time_steps, height, width, 1))
```

```python
time_steps
```

```python
for i in range(time_steps):
    # Assuming that the third value in precipitation_data[i] is the precipitation value
    tp_value = precipitation_data[i][2]  # Index 2 corresponds to the 'tp' value in your data
    dynamic_data[i, :, :, 0] = tp_value  # Broadcast the precipitation value across the grid
```

```python
   # Normalize dynamic data
dynamic_data_reshaped = dynamic_data.reshape(-1, dynamic_data.shape[-1])
dynamic_data_reshaped_scaled = scaler.fit_transform(dynamic_data_reshaped)
dynamic_data = dynamic_data_reshaped_scaled.reshape(dynamic_data.shape)
```

```python
  print(f"Dynamic data reshaped and normalized with shape {dynamic_data.shape}")
```

```python
# Visualization of dynamic data (first time step)
show_band(dynamic_data[0, :, :, 0], "Precipitation Data (First Time Step)", cmap='Blues')
```

```python
# Step 3: Load and Preprocess Flood Map Data (Target) with Masking
def load_flood_maps(folder_path):
    flood_maps = []
    for filename in sorted(os.listdir(folder_path)):  # Ensure files are sorted by time
        if filename.endswith('.tif'):
            with rasterio.open(os.path.join(folder_path, filename)) as src:
                flood_map = src.read(1)  # Read the first band
                flood_map = np.ma.masked_equal(flood_map, src.nodata)  # Mask nodata values
                flood_maps.append(flood_map)
                print(f"Loaded flood map {filename} with shape {flood_map.shape}")
    flood_maps = np.ma.stack(flood_maps, axis=0)  # Shape: (time_steps, height, width)
    
    # Show the first time step of the flood map
    show_band(flood_maps[0, :, :], title="Flood Map Data (First Time Step)", cmap='Blues')
    
    return flood_maps
```

```python
# Path to the folder containing flood map rasters
flood_map_folder = 'calraster/clipped_rasters'
```

```python
# Load flood maps with masking and visualization
flood_maps = load_flood_maps(flood_map_folder)
```

```python
# Normalize flood maps
flood_maps_reshaped = flood_maps.reshape(-1, 1)
flood_maps_reshaped_scaled = scaler.fit_transform(flood_maps_reshaped)
flood_maps = flood_maps_reshaped_scaled.reshape(flood_maps.shape)
```

```python
# Reshape for model input (adding a channel dimension)
flood_maps = flood_maps[..., np.newaxis]  # Shape: (time_steps, height, width, 1)
```

```python
print(f"Flood maps normalized with shape {flood_maps.shape}")
```

```python
# Step 4: Prepare Data for ConvLSTM Model
# Combine static and dynamic data
static_data_tiled = np.tile(static_data, (time_steps, 1, 1, 1))  # Replicate static data across time steps
```

```python
# Combine static and dynamic features
combined_data = np.concatenate([static_data_tiled, dynamic_data], axis=-1)  # Shape: (time_steps, height, width, channels)
```

```python
X = combined_data.reshape(1, time_steps, height, width, combined_data.shape[-1])  # Adding batch dimension
y = flood_maps.reshape(1, *flood_maps.shape)  # Adding batch dimension
```

```python
print(f"Training data shape: X: {X.shape}, y: {y.shape}")
```

```python

# print(f"Training data shape: X_train: {X_train.shape}, y_train: {y_train.shape}")
# print(f"Testing data shape: X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
model = Sequential([
    ConvLSTM2D(filters=64, kernel_size=(3, 3), input_shape=X.shape[1:], padding='same', return_sequences=True),
    BatchNormalization(),
    ConvLSTM2D(filters=64, kernel_size=(3, 3), padding='same', return_sequences=True),
    BatchNormalization(),
    ConvLSTM2D(filters=64, kernel_size=(3, 3), padding='same', return_sequences=False),
    BatchNormalization(),
    Reshape((1, height, width, 64)),  # Reshape to add the extra dimension required by Conv3D
    Conv3D(filters=1, kernel_size=(3, 3, 3), activation='linear', padding='same')
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X, y, epochs=20, batch_size=8)
```

```python
# Plot training history
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
