```python
import numpy as np
import rasterio
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

def load_raster(file_path, title):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        masked_data = np.ma.masked_equal(data, src.nodata)  # Mask nodata values
    print(f"Loaded {title} with shape {masked_data.shape}")
    return masked_data

def show_band(data, title='', size=(10, 10), cmap='terrain', colorbar=True):
    plt.figure(figsize=size)
    plt.imshow(data, cmap=cmap)
    plt.title(title)
    if colorbar:
        plt.colorbar()
    plt.axis('off')
    plt.show()

# Load the elevation and flood TIFF files
elevation_path = 'snapped_data/snapped_DEM.tif'
flood_map_path = 'calraster/clipped_rasters/clipped_Sim7_2BaseDefault_2D_overland_elements_095.shp.tif'

elevation_data = load_raster(elevation_path, "Elevation Data")
flood_data = load_raster(flood_map_path, "Flood Data")

# Visualize the elevation and flood data
show_band(elevation_data, title='Elevation Data', cmap='terrain')
show_band(flood_data, title='Flood Data', cmap='Blues')

# Reshape data to 2D
height, width = elevation_data.shape
elevation_reshaped = elevation_data.reshape(-1, 1)
flood_reshaped = flood_data.reshape(-1, 1)

# Normalize the data
scaler = StandardScaler()
elevation_scaled = scaler.fit_transform(elevation_reshaped)
flood_scaled = scaler.fit_transform(flood_reshaped)

# Train a Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(elevation_scaled, flood_scaled.ravel())

# Predict the flood map using the entire elevation data
flood_pred_scaled = rf_model.predict(elevation_scaled)

# Reshape to original dimensions
flood_pred = scaler.inverse_transform(flood_pred_scaled.reshape(-1, 1)).reshape(height, width)

# Visualize the predicted flood map
show_band(flood_pred, title='Predicted Flood Map from Elevation Data', cmap='Blues')

# Optional: Calculate the Mean Squared Error
mse = mean_squared_error(flood_scaled, flood_pred_scaled)
print(f"Mean Squared Error: {mse}")
```
