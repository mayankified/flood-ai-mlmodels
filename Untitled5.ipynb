```python
import numpy as np
import rasterio
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Input
from sklearn.metrics import mean_squared_error

def load_raster(file_path, title):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        masked_data = np.ma.masked_equal(data, src.nodata)  # Mask nodata values
    print(f"Loaded {title} with shape {masked_data.shape}")
    return masked_data

def show_band(data, title='', size=(10, 10), cmap='terrain', colorbar=True):
    plt.figure(figsize=size)
    plt.imshow(data, cmap=cmap)
    plt.title(title)
    if colorbar:
        plt.colorbar()
    plt.axis('off')
    plt.show()

def extract_patches_with_stride(data, patch_size, stride):
    height, width, channels = data.shape
    
    patches = []
    for i in range(0, height - patch_size + 1, stride):
        for j in range(0, width - patch_size + 1, stride):
            patch = data[i:i+patch_size, j:j+patch_size, :]
            patches.append(patch)
    
    patches = np.array(patches)
    return patches, height, width

def reconstruct_from_patches(patches, image_shape, patch_size, stride):
    height, width = image_shape
    reconstructed_image = np.zeros((height, width))
    patch_count = np.zeros((height, width))

    patch_idx = 0
    for i in range(0, height - patch_size + 1, stride):
        for j in range(0, width - patch_size + 1, stride):
            reconstructed_image[i:i+patch_size, j:j+patch_size] += patches[patch_idx]
            patch_count[i:i+patch_size, j:j+patch_size] += 1
            patch_idx += 1

    # Handle zero `patch_count` by replacing zeros with ones to avoid division by zero
    patch_count[patch_count == 0] = 1

    # Average the overlapping areas
    reconstructed_image /= patch_count
    
    return reconstructed_image

# Load the elevation and flood TIFF files
elevation_path = 'snapped_data/snapped_DEM.tif'
flood_map_path = 'calraster/clipped_rasters/clipped_Sim7_2BaseDefault_2D_overland_elements_095.shp.tif'

elevation_data = load_raster(elevation_path, "Elevation Data")
flood_data = load_raster(flood_map_path, "Flood Data")

# Visualize the elevation and flood data
show_band(elevation_data, title='Elevation Data', cmap='terrain')
show_band(flood_data, title='Flood Data', cmap='terrain')

# Normalize the data
scaler = StandardScaler()
elevation_scaled = scaler.fit_transform(elevation_data.reshape(-1, 1)).reshape(elevation_data.shape)
flood_scaled = scaler.fit_transform(flood_data.reshape(-1, 1)).reshape(flood_data.shape)

# Reshape the data for CNN input (add a channel dimension)
elevation_scaled = elevation_scaled[..., np.newaxis]
flood_scaled = flood_scaled[..., np.newaxis]

# Extract patches with striding
patch_size = 64
stride = 32

elevation_patches, height, width = extract_patches_with_stride(elevation_scaled, patch_size, stride)
flood_patches, _, _ = extract_patches_with_stride(flood_scaled, patch_size, stride)

# Define a CNN model for patch-based training
model = Sequential([
    Input(shape=(patch_size, patch_size, 1)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    UpSampling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    UpSampling2D((2, 2)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Conv2D(1, (3, 3), activation='linear', padding='same')  # Output layer
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model on the patches
history = model.fit(elevation_patches, flood_patches, epochs=50, batch_size=8)

# Predict the flood map patches using the CNN
flood_pred_patches = model.predict(elevation_patches)

# Reconstruct the full flood map from predicted patches
flood_pred_scaled = reconstruct_from_patches(flood_pred_patches, (height, width), patch_size, stride)

# Inverse transform the scaled predictions to original scale
flood_pred = scaler.inverse_transform(flood_pred_scaled.reshape(-1, 1)).reshape(height, width)

# Visualize the reconstructed predicted flood map
show_band(flood_pred, title='Reconstructed Predicted Flood Map from Elevation Data (CNN)', cmap='terrain')

# Optional: Calculate the Mean Squared Error
mse = mean_squared_error(flood_scaled.reshape(-1, 1), flood_pred_scaled.reshape(-1, 1))
print(f"Mean Squared Error: {mse}")

```
