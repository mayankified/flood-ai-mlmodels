```python
import numpy as np
import rasterio
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Path to your raster file (e.g., a .tif file)
raster_file = 'calraster/floodimg.tif'

# Open the raster file
with rasterio.open(raster_file) as src:
    # Read the single band
    band = src.read(1)  # Since there is only one band
    
    # Mask the nodata values
    flood_img = np.ma.masked_equal(band, src.nodata)

    # Flatten the masked array for scaling
    band_reshaped = flood_img.compressed().reshape(-1, 1)
    
    # Initialize the StandardScaler
    scaler = StandardScaler()

    # Fit and transform the band
    band_scaled = scaler.fit_transform(band_reshaped)

    # Fill the scaled values back into the original masked array
    flood_img_scaled = flood_img.copy()
    flood_img_scaled.data[~flood_img.mask] = band_scaled.flatten()
    
    # Apply the mask again to the scaled image
    flood_img_scaled = np.ma.masked_array(flood_img_scaled, mask=flood_img.mask)

# Display the band before and after scaling for comparison
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Band")
plt.imshow(flood_img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Scaled Band")
plt.imshow(flood_img_scaled, cmap='gray')
plt.axis('off')

plt.show()
```
