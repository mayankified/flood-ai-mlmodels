```python
import numpy as np
import rasterio
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Input
from sklearn.metrics import mean_squared_error

def load_raster(file_path, title):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        masked_data = np.ma.masked_equal(data, src.nodata)  # Mask nodata values
    print(f"Loaded {title} with shape {masked_data.shape}")
    return masked_data

def show_band(data, title='', size=(10, 10), cmap='terrain', colorbar=True):
    plt.figure(figsize=size)
    plt.imshow(data, cmap=cmap)
    plt.title(title)
    if colorbar:
        plt.colorbar()
    plt.axis('off')
    plt.show()

# Load the elevation and flood TIFF files
elevation_path = 'snapped_data/snapped_DEM.tif'
flood_map_path = 'calraster/clipped_rasters/clipped_Sim7_2BaseDefault_2D_overland_elements_095.shp.tif'

elevation_data = load_raster(elevation_path, "Elevation Data")
flood_data = load_raster(flood_map_path, "Flood Data")

# Visualize the elevation and flood data
show_band(elevation_data, title='Elevation Data', cmap='terrain')
show_band(flood_data, title='Flood Data', cmap='terrain')

# Get the height and width of the images
height, width = elevation_data.shape

# Normalize the data
scaler = StandardScaler()
elevation_scaled = scaler.fit_transform(elevation_data.reshape(-1, 1)).reshape(elevation_data.shape)
flood_scaled = scaler.fit_transform(flood_data.reshape(-1, 1)).reshape(flood_data.shape)

# Reshape the data for CNN input (add a channel dimension)
elevation_scaled = elevation_scaled[..., np.newaxis]
flood_scaled = flood_scaled[..., np.newaxis]

# Define a CNN model with explicit input layer
model = Sequential([
    Input(shape=(height, width, 1)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    UpSampling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    UpSampling2D((2, 2)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Conv2D(1, (3, 3), activation='linear', padding='same')  # Output layer
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model without validation split
history = model.fit(elevation_scaled[np.newaxis, ...], flood_scaled[np.newaxis, ...], epochs=50, batch_size=1)

# Predict the flood map using the CNN
flood_pred_scaled = model.predict(elevation_scaled[np.newaxis, ...])

# Reshape to remove the batch dimension and channel
flood_pred_scaled = flood_pred_scaled[0, ..., 0]

# Inverse transform the scaled predictions to original scale
flood_pred = scaler.inverse_transform(flood_pred_scaled.reshape(-1, 1)).reshape(height, width)

# Visualize the predicted flood map
show_band(flood_pred, title='Predicted Flood Map from Elevation Data (CNN)', cmap='terrain')

# Optional: Calculate the Mean Squared Error
mse = mean_squared_error(flood_scaled.reshape(-1, 1), flood_pred_scaled.reshape(-1, 1))
print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_focused_band(data, title='', size=(10, 10), cmap='terrain_r', vmin=None, vmax=None):
    """
    Plots the data with a focused color range, hiding values outside the specified range.

    Parameters:
    - data: The 2D array of data to plot.
    - title: The title of the plot.
    - size: The size of the plot (default is 10x10).
    - cmap: The colormap to use (default is 'terrain_r' for reversed terrain).
    - vmin: The minimum value for the color range (values below this will be hidden).
    - vmax: The maximum value for the color range (values above this will be hidden).

    Returns:
    - None
    """
    # Mask values outside the specified range
    masked_data = np.ma.masked_outside(data, vmin, vmax)
    
    plt.figure(figsize=size)
    plt.imshow(masked_data, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Example Usage:
# Assuming 'flood_pred' is your data and you want to focus on the dark brown range

# Set vmin and vmax to focus on the desired color range
vmin = 1000    # Adjust this to the lower bound of the desired range
vmax = 5000    # Adjust this to the upper bound of the desired range

plot_focused_band(flood_pred, title='Predicted Flood Map (Focused on Specific Range)', vmin=vmin, vmax=vmax)
```
