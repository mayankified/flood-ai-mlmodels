```python
import numpy as np
import pandas as pd
import rasterio
import os
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, UpSampling2D, MaxPooling2D, Input
import matplotlib.pyplot as plt

# Function to plot the data
def show_band(data, title='', size=(10, 10), cmap='viridis', vmin=None, vmax=None):
    plt.figure(figsize=size)
    plt.imshow(data, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.title(title)
    plt.colorbar()
    plt.axis('off')
    plt.show()

# Function to load and mask raster data
def load_static_raster(file_path, title):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        masked_data = np.ma.masked_equal(data, src.nodata)  # Mask nodata values
    print(f"Loaded {title} with shape {masked_data.shape}")
    show_band(masked_data, title=title)  # Show the raster with color map
    return masked_data

# Paths to your static datasets
elevation_path = 'snapped_data/snapped_DEM.tif'
landuse_path = 'snapped_data/snapped_landuse.tif'
flood_map_path = 'calraster/clipped_rasters/clipped_Sim7_2BaseDefault_2D_overland_elements_095.shp.tif'

# Load static rasters with masking and visualization
elevation_data = load_static_raster(elevation_path, "Elevation Data")
landuse_data = load_static_raster(landuse_path, "Landuse Data")
flood_data = load_static_raster(flood_map_path, "Flood Data")

# Step 1: Visualize the Original Flood Data Without Scaling
show_band(flood_data, title="Original Flood Data", cmap='viridis')

# Check the min and max values of the original flood data
print(f"Original Flood Data min: {np.min(flood_data)}, max: {np.max(flood_data)}")

# Step 2: Examine the Histogram of the Original Data
plt.figure(figsize=(8, 6))
plt.hist(flood_data.ravel(), bins=50, color='blue', alpha=0.7)
plt.title("Histogram of Original Flood Data")
plt.xlabel("Flood Depth Value")
plt.ylabel("Frequency")
plt.show()

# Decision: Apply MinMaxScaler or Skip Scaling
# If you decide to scale:
height, width = flood_data.shape
flood_data_reshaped = flood_data.reshape(-1, 1)

# Apply MinMaxScaler
scaler = MinMaxScaler()
flood_data_scaled = scaler.fit_transform(flood_data_reshaped)

# Reshape flood data to 3D (height, width, 1)
flood_data_scaled_3d = flood_data_scaled.reshape(height, width, 1)

# Visualize the scaled flood data
show_band(flood_data_scaled_3d[:, :, 0], title="Scaled Flood Data", cmap='viridis')

# If you decide to skip scaling:
# flood_data_scaled_3d = flood_data.reshape(height, width, 1)

# Step 3: Normalize the Static Data
static_data = np.ma.stack([elevation_data, landuse_data], axis=-1)
static_data_reshaped = static_data.reshape(-1, static_data.shape[-1])

# Normalize static data
static_data_reshaped_scaled = scaler.fit_transform(static_data_reshaped)
static_data = static_data_reshaped_scaled.reshape(static_data.shape)

# Visualize the normalized static data
show_band(static_data[..., 0], "Elevation Data (Normalized)", cmap='terrain')
show_band(static_data[..., 1], "Landuse Data (Normalized)", cmap='RdYlBu')

# Load dynamic data
def load_dynamic_csv(file_path):
    df = pd.read_csv(file_path)
    return df['tp'].values

precipitation_path = 'calraster/precipitation_data.csv'
precipitation_values = load_dynamic_csv(precipitation_path)
precipitation_value = precipitation_values[95]

precipitation_data = np.full((elevation_data.shape[0], elevation_data.shape[1], 1), precipitation_value)

# Normalize precipitation data
precipitation_data_reshaped = precipitation_data.reshape(-1, 1)
precipitation_data_reshaped_scaled = scaler.fit_transform(precipitation_data_reshaped)
precipitation_data = precipitation_data_reshaped_scaled.reshape(precipitation_data.shape)

# Visualize the scaled precipitation data
show_band(precipitation_data[:, :, 0], title="Scaled Precipitation Data", cmap='viridis')

# Combine static and dynamic data
combined_data = np.concatenate([static_data, precipitation_data], axis=-1)

# Visualize the combined data
print(f"Combined data shape: {combined_data.shape}")

# Step 4: Extract Patches
def extract_patches_with_stride(data, patch_size, stride):
    height, width, channels = data.shape
    patches = []
    for i in range(0, height - patch_size + 1, stride):
        for j in range(0, width - patch_size + 1, stride):
            patch = data[i:i+patch_size, j:j+patch_size, :]
            patches.append(patch)
    return np.array(patches)

patch_size = 64
stride = 32

combined_patches = extract_patches_with_stride(combined_data, patch_size, stride)

# Extract patches from the 3D flood data
flood_map_patches = extract_patches_with_stride(flood_data_scaled_3d, patch_size, stride)

# Check the first few patches
print(f"First combined patch min: {np.min(combined_patches[0])}, max: {np.max(combined_patches[0])}")
print(f"First flood patch min: {np.min(flood_map_patches[0])}, max: {np.max(flood_map_patches[0])}")

# Step 5: Build and Train the CNN Model
model = Sequential([
    Input(shape=(patch_size, patch_size, combined_patches.shape[-1])),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    UpSampling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    UpSampling2D((2, 2)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Conv2D(1, (3, 3), activation='linear', padding='same')
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(combined_patches, flood_map_patches, epochs=50, batch_size=8, validation_split=0.1)

# Plot training history
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Step 6: Reconstruct the Flood Map
def reconstruct_from_patches(patches, image_shape, patch_size, stride):
    height, width = image_shape
    reconstructed_image = np.zeros((height, width))
    patch_count = np.zeros((height, width))

    patch_idx = 0
    for i in range(0, height - patch_size + 1, stride):
        for j in range(0, width - patch_size + 1, stride):
            reconstructed_image[i:i+patch_size, j:j+patch_size] += patches[patch_idx, :, :, 0]
            patch_count[i:i+patch_size, j:j+patch_size] += 1
            patch_idx += 1

    patch_count[patch_count == 0] = 1
    reconstructed_image /= patch_count

    return reconstructed_image

predicted_patches = model.predict(combined_patches)
reconstructed_flood_map = reconstruct_from_patches(predicted_patches, flood_data.shape, patch_size, stride)

# Apply inverse transform to get the original scale
reconstructed_flood_map_reshaped = reconstructed_flood_map.reshape(-1, 1)
reconstructed_flood_map_original_scale = scaler.inverse_transform(reconstructed_flood_map_reshaped)
reconstructed_flood_map_original_scale = reconstructed_flood_map_original_scale.reshape(flood_data.shape)

# Check the min and max values of the reconstructed map
print(f"Reconstructed Flood Map (original scale) min: {np.min(reconstructed_flood_map_original_scale)}, max: {np.max(reconstructed_flood_map_original_scale)}")

# Adjust vmin and vmax for better visualization if needed
vmin, vmax = np.min(flood_data), np.max(flood_data)

# Plot the reconstructed flood map in the original scale
show_band(reconstructed_flood_map_original_scale, title='Reconstructed Flood Map (Original Scale)', size=(10, 10), cmap='viridis', vmin=vmin, vmax=vmax)
```
